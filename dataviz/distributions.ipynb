```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
print("Setup Complete")
```

```python
iris_data = pd.read_csv('~\JProjects\kaggle\data\iris.csv', index_col='Id')

iris_data.head()
```

## Histograms

```python
sns.histplot(iris_data.iloc[:, [2]])
```

## Density plots

***Kernel density estimate (KDE)*** plot. You can think of it as a ***smoothed histogram***.

```python
sns.kdeplot(data=iris_data['Petal Length (cm)'], fill=True) # fill == shade
```

```python
sns.kdeplot(x=iris_data['Petal Length (cm)'], y=iris_data['Sepal Width (cm)'], fill=True)
```

 We can create a two-dimensional (2D) KDE plot with the ***sns.jointplot*** command.

```python
sns.jointplot(x=iris_data['Petal Length (cm)'], y=iris_data['Sepal Width (cm)'])
```

```python
sns.jointplot(x=iris_data['Petal Length (cm)'], y=iris_data['Sepal Width (cm)'], kind='kde', fill=True)
```

In the plot Above, the color-coding shows us how likely we are to see different combinations of sepal width and petal length, where darker parts of the figure are more likely.

## Color-coded plots

For the next part of the tutorial, we'll create plots to understand differences between the species.

We can create three different histograms (one for each species) of petal length by using the sns.histplot command (as above).
- data= provides the name of the variable that we used to read in the data
- x= sets the name of column with the data we want to plot
- hue= sets the column we'll use to split the data into different histograms


```python
sns.histplot(data=iris_data, x='Petal Length (cm)', hue='Species')
plt.title("Histogram of Petal Lengths, by Species")
```

```python
sns.kdeplot(data=iris_data, x='Petal Length (cm)', hue='Species', fill=True)
plt.title("Histogram of Petal Lengths, by Species")
```
