```python
import pandas as pd


url = 'https://raw.githubusercontent.com/chrisalbon/simulated_datasets/master/titanic.csv'

data = pd.read_csv(url)
```

```python
for name in data['Name'][:2]:
  print(name.upper())
```



---



To apply a built-in or user functions to each elements in a column we use the *apply* method.

```python
def uppercase(x):
  return x.upper()


lowercase = lambda x: x.lower()
```

```python
data['Name'].apply(uppercase)[:3]
```

```python
data['Name'].apply(lowercase)[:3]
```

```python
data.groupby('Sex').apply(lambda x : x .count())
```
