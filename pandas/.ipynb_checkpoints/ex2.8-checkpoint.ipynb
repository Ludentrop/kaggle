For rows grouping by time intervals we use the *resample* method. It requires the data frame has datetime type of indexes.
As default the method a right side label - the last time label. We can control it by the *label* parameter.

```python
time_index = pd.date_range('06/06/2017', periods=100000, freq='30S')

data = pd.DataFrame(index=time_index)

data['Sale_Amount'] = np.random.randint(1, 10, 100000)
```

```python
pd.Series(pd.date_range("1/1/2011", freq="M", periods=3))
```

```python
data.resample('W').sum()  # by week
```

```python
data.resample('M').sum()  # by month
```

```python
data.resample('M', label='left').sum()
```
