```python
import pandas as pd


url = 'https://raw.githubusercontent.com/chrisalbon/simulated_datasets/master/titanic.csv'

data = pd.read_csv(url)
```

```python
data.head(5)
```

```python
data.drop('Age', axis=1).head(5)
```

```python
data.drop(['Age', 'PClass'], axis=1).head(5)
```

```python
data.drop(data.columns[0], axis=1).head(5)  # if a column is not nammed
```

In order to delete duplicates in the data frame we use *drop_duplicates* that as default saves first unique values but we can control it by *keep* parameter. Usually we search duplicates in a subset of the data frame and for the reason we use *subset* parameter of the method.

```python
data.drop_duplicates(subset=['Sex'], keep='last')
```

```python
data.head(5)
```

```python
data.duplicated(subset=['Sex']).head(5)
```
